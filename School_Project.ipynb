```python

# 2. Compute per‐state metrics
state_stats = (
    df
    .groupby('State')
    .agg(
        avg_satisfaction=('Satisfaction Rate', 'mean'),
        churn_rate=('Customer Churn Status', lambda x: (x == 'Yes').mean() * 100)
    )
    .sort_values('avg_satisfaction', ascending=False)
)

# 3. Plot
plt.figure(figsize=(14, 6))
ax1 = plt.gca()
ax2 = ax1.twinx()

# — Bar chart of average satisfaction (steelblue) —
sns.barplot(
    x=state_stats.index,
    y=state_stats['avg_satisfaction'],
    color='steelblue',
    ax=ax1
)

# — Line plot of churn rate (crimson) —
sns.lineplot(
    x=state_stats.index,
    y=state_stats['churn_rate'],
    marker='o',
    sort=False,
    color='crimson',
    ax=ax2
)

# 4. Styling
ax1.set_xlabel('State', fontsize=12)
ax1.set_ylabel('Avg. Satisfaction Rate', fontsize=12, color='steelblue')
ax2.set_ylabel('Churn Rate (%)', fontsize=12, color='crimson')
ax1.set_title('State-by-State: Avg. Satisfaction vs. Churn Rate', fontsize=14)

# Rotate x‐axis labels for readability
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

# Grid only on the left (bars) axis
ax1.grid(axis='y', linestyle='--', alpha=0.5)
ax2.grid(False)

plt.tight_layout()
plt.show()
```

### Total revenue by Age

```python
bins   = [17, 25, 35, 45, 55, 65, 100]
labels = ['18–25', '26–35', '36–45', '46–55', '56–65', '65+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_revenue = (
    df
    .groupby('Age Group', observed=False)['Total Revenue']
    .sum()
    .reset_index()
)
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Age Group',
    y='Total Revenue',
    data=age_revenue,
    palette='viridis',
    hue = 'Age Group'
  
)
plt.title('Total Revenue by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
ct = pd.crosstab(df['Gender'], df['Customer Churn Status'])

# 3. Plot a stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Keep track of the bottom of each stack
bottom = [0] * len(ct)

for status in ct.columns:
    ax.bar(
        ct.index,           # x-axis categories (Male/Female)
        ct[status],         # heights for this status
        bottom=bottom,      # stack on top of previous bars
        label=status        # legend label
    )
    # update bottom for next status
    bottom = bottom + ct[status].values

# 4. Customize plot
ax.set_title('Gender Distribution across Churn Status')
ax.set_xlabel('Gender')
ax.set_ylabel('Number of Customers')
ax.legend(title='Churn Status')
plt.tight_layout()

# 5. Show it
plt.show()
```

Similar Retention Across Genders:

Both males and females have over 300 customers who stayed (did not churn).

This suggests MTN is equally successful at retaining both male and female customers.

Similar Churn Behavior:

Over 100 customers in each gender group churned.

The churn rate appears comparable across genders, implying gender is not a strong differentiator in predicting churn in this dataset.

Balanced Gender Impact:

The distribution hints that gender alone may not significantly influence churn decisions.

Any churn-related interventions likely won’t need to be gender-specific.
